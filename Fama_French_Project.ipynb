```python
#Bỏ các cột không sử dụng được như những cột chứa ERROR, những quỹ đầu tư (ETF), những cột ghi nhận doanh nghiệp bị delist 
#và những cột Unnamed
def delete_unsuable_column(DF):
    DF = DF[DF.columns.drop(list(DF.filter(regex='Unnamed:')))]
    for i in DF.columns:
        if 'ERROR' in i or 'ETF' in i or 'DELIST' in i:
            DF = DF.drop(i,axis=1)
    return DF
df_price=delete_unsuable_column(df_price)
df_bookvalue=delete_unsuable_column(df_bookvalue)
df_marketvalue=delete_unsuable_column(df_marketvalue)
```

```python
#Xóa tất cả những hàng mà tất cả các giá trị đều là NA
def drop_row_na(DF):
    l=[]
    for i in range(DF.shape[0]):
        if DF.iloc[i,:].isnull().sum()==DF.shape[1]:
            l.append(i)
    DF=DF.drop(DF.index[l])
    return DF
df_price=drop_row_na(df_price)
df_bookvalue=drop_row_na(df_bookvalue)
df_marketvalue=drop_row_na(df_marketvalue)
```

```python
#Xóa tất cả những cột mà tất cả các giá trị đều là NA
def drop_column_na(DF):
    for i in DF.columns:
        if DF[i].isnull().sum().sum()==DF.shape[0]-1:
            DF=DF.drop(i,axis=1)
    return DF
df_price=drop_column_na(df_price)
df_bookvalue=drop_column_na(df_bookvalue)
df_marketvalue=drop_column_na(df_marketvalue)
```

```python
def Fama_French(i):
    l=[]
    #Tạo DataFrame gồm 3 cột Ri, B/M, Size từ các bảng dữ liệu
    df=pd.DataFrame({'Ri':df_price_r.iloc[i,:].values.tolist()
                  ,'B/P':df_bookvalue.iloc[i,:]/df_price.iloc[i,:]
                 ,'Size':df_marketvalue.iloc[i,:]})
    df=df.dropna()
    #Sort DataFrame theo size
    df=df.sort_values(by='Size')
    #Chia nhóm Big,Small
    df_small=df[df['Size']<df['Size'].values.mean()]
    df_big=df[df['Size']>=df['Size'].values.mean()]
    #Sort 2 nhóm Big,Small theo B/M
    df_small=df_small.sort_values(by='B/P')
    df_big=df_big.sort_values(by='B/P')
    #Chia L,M,H cho nhóm Small
    df_SL=df_small.iloc[:int(df_small.shape[0]*0.3)+1,:]
    df_SM=df_small.iloc[int(df_small.shape[0]*0.3)+1:int(df_small.shape[0]*0.3)+1+int(df_small.shape[0]*0.4),:]
    df_SH=df_small.iloc[int(df_small.shape[0]*0.3)+1+int(df_small.shape[0]*0.4):,:]
    #Chia L,M,H cho nhóm Big
    df_BL=df_big.iloc[:int(df_big.shape[0]*0.3)+1,:]
    df_BM=df_big.iloc[int(df_big.shape[0]*0.3)+1:int(df_big.shape[0]*0.3)+1+int(df_big.shape[0]*0.4),:]
    df_BH=df_big.iloc[int(df_big.shape[0]*0.3)+1+int(df_big.shape[0]*0.4):,:]
    #Tính tỷ suất lợi nhuận trung bình cho các nhóm
    R_SL=df_SL['Ri'].values.mean()
    R_SM=df_SM['Ri'].values.mean()
    R_SH=df_SH['Ri'].values.mean()
    R_BL=df_BL['Ri'].values.mean()
    R_BM=df_BM['Ri'].values.mean()
    R_BH=df_BH['Ri'].values.mean()
    #Tính R_SMB và R_HML
    R_SMB = (1/3)*(R_SL+R_SM+R_SH)-(1/3)*(R_BL+R_BM+R_BH)
    R_HML = (1/2)*(R_BH+R_SH)-(1/2)*(R_BL+R_SL)
    #Tính tỷ suất lợi nhuận trung bình của các cổ phiếu trong tháng
    Ri=(R_SL+R_SM+R_SH+R_BL+R_BM+R_BH)/6
    l.append(Ri)
    l.append(R_SMB)
    l.append(R_HML)
    return l
```

```python
#Cắt bảng dữ liệu Rf,Rm theo khoảng thời gian 10 năm đã lựa chọn ở trên (2010-09/2021)
df_R=df_R.loc['01/2010':'09/2021',:]
df_R
```

```python
#Tạo những list các giá trị Ri, R_SMB và R_HML đã tính được
list_Ri=[]
list_R_SMB=[]
list_R_HML=[]
for i in range(len(df_price_r)):
    list_Ri.append(Fama_French(i)[0])
    list_R_SMB.append(Fama_French(i)[1])
    list_R_HML.append(Fama_French(i)[2])
```

```python
#Thêm những cột dữ liệu cần thiết cho mô hình hồi quy
df_R['Ri']= list_Ri
df_R['Ri-Rf']= df_R['Ri']-df_R['Rf']
df_R['Rm-Rf']= df_R['VNINDEX Rm']- df_R['Rf']
df_R['Rsmb']= list_R_SMB
df_R['Rhml']= list_R_HML
df_R=df_R.dropna()
df_R
```

```python
#Triển khai hồi quy
y= df_R['Ri-Rf']
x= df_R[['Rm-Rf','Rsmb','Rhml']]
x= sm.add_constant(x)
x, y= np.array(x), np.array(y)
model = sm.OLS(y,x).fit()
print(model.summary())
```
